### 2. import csv data from url

```python
import requests
from io import StringIO

url = "https://raw.githubusercontent.com/cs109/2014_data/master/countries.csv"
headers = {"User-Agent": "Mozilla/5.0 (Macintosh; Intel Mac OS X 10.14; rv:66.0) Gecko/20100101 Firefox/66.0"}

req = requests.get(url, headers=headers)
data = StringIO(req.text)

df = pd.read_csv(data)

df.sample()
```

### 3. tab seperated value(tsv)

if you have tab seperated value instead of comma seperated then you have to use 'sep' parameter as '\t'

```python
pd.read_csv('movie_titles_metadata.tsv',sep='\t')
```

### 4. missing header

as you can see there is a problem of of missing a headers in column and our data set using first row of data as a header for that problem we can use 'names' parameter 

```python
df = pd.read_csv('movie_titles_metadata.tsv',sep='\t',names=['sno','name','release_year','rating','votes','genres'])
df.sample()
```

### 5. create Index column 

some times in our dataset we can see we have index of column so in this condition we can replace auto generated index with our data index 

```python
df = pd.read_csv('aug_train.csv',index_col='enrollee_id')

df.sample()
```

### 6. header as a row

sometimes you can face a problem where your header use as a column so macking the header we can use 'header' parameter

```python
df = pd.read_csv('test.csv',sep='\t',header=1)

df.head()
```

### 7. columns which need

sometimes we dont need all the column instead of all we need some column so in this situation we can use 'use_cols' parameter for getting requed parameters

```python
df = pd.read_csv('aug_train.csv',usecols=['enrollee_id','gender','education_level'])

df.sample()
```

### 8. skiprows/nrow parameters

sometimes we have to skip some value in this situation we can use skiprows parameter and we can also have nrows instead of getting all record we can get some record and operate them

```python
df = pd.read_csv('aug_train.csv',skiprows=[0,2])

df.sample()
```

```python
df = pd.read_csv('aug_train.csv',nrows = 100)

df.sample()
```

### 9. Encoding

some times we gets a dataset of different encoding in this situation we can use 2 approach 1st change encoding by using online tools and by finding encoding of data pass that encoding to 'encoding' parameter

```python
df = pd.read_csv('zomato.csv',encoding='latin-1')

df.head(2)
```

### 10. skip bad lines

sometimes we get a situation in which some of rows have a extra values or extra column in this situation our pandas show error like "Expected n fields in line 6452, saw n+1" the solution is we can skip these datas by using 'error_bad_lines' parameter

```python
pd.read_csv('BX-Books.csv', sep=';', encoding="latin-1",error_bad_lines=False)
```

### 11. Change data types of column

```python
df = pd.read_csv("aug_train.csv")

before_type = type(df['target'][0])

df = pd.read_csv('aug_train.csv',dtype={'target':int})

after_type = type(df['target'][0])

print('data type before ',before_type, '\ndata type after ', after_type)
```

### 12. Handling Dates

sometimes we get data set of date column in a string formate for converting then into date we can use 'parse_dates' parameter

before date column type = object 

after date column type = datetime64[ns]

```python
pd.read_csv('ipl_match.tsv',sep = '\t',parse_dates=['date']).info()
```

### 13. Convertors

this is a preprocessing step in this we can change column content into other content by creating a fuction
in sort we can say applying any fuction to a column 

" converters={'column_name':fuction} "

```python
def rename(name):
    if name == "Royal Challengers Bangalore":
        return "RCB"
    else:
        return name
```

```python
pd.read_csv('ipl_match.tsv',sep = '\t',converters={'team1':rename})
```

### 14.nan value

sometimes in our data set we have to put nan on certain condition in this situation we can use 'na_value' parameter


put nan where we have male in dataset

```python
pd.read_csv('aug_train.csv',na_values=['Male',])
```

### 15. Loading a huge dataset in chunks

sometimes we have large amount of dataset to applying function to dataset its difficult to manage all data at a same times so we can mack a chunks and divid data into small parts

```python
dfs = pd.read_csv('aug_train.csv',chunksize=5000)
```

```python
for chunks in dfs:
    print(chunks.shape)
```
